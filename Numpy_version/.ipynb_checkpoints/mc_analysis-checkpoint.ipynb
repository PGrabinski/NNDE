```python
import pandas as pd
import os
```

```python
directory = next(os.walk('.'))
```

```python
for file in directory[2]:
    parts = file.split('.')
    if parts[-1] == 'csv':
        table = pd.read_csv(file, sep='\t')
        print(parts[0])
        print(table.dropna().describe().applymap(lambda x: '%E' % x))
```
